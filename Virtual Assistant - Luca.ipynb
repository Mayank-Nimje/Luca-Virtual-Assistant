```python
engine=pyttsx3.init('sapi5')
voices=engine.getProperty('voices')
engine.setProperty('voice','voices[0].id')
```

```python
def speak(text):
    engine.say(text)
    engine.runAndWait()
```

```python
def wishMe():
    hour=datetime.datetime.now().hour
    if hour>=0 and hour<12:
        speak("Hello,Good Morning, I am Luca an intelligent virtual assisstant")
        print("Hello,Good Morning, I am Luca an intelligent virtual assisstant")
    elif hour>=12 and hour<18:
        speak("Hello,Good Afternoon, I am Luca an intelligent virtual assisstant")
        print("Hello,Good Afternoon, I am Luca an intelligent virtual assisstant")
    else:
        speak("Hello,Good Evening, I am Luca an intelligent virtual assisstant")
        print("Hello,Good Evening, I am Luca an intelligent virtual assisstant")
```

```python
def takeCommand():
    r=sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        audio=r.listen(source)

        try:
            statement=r.recognize_google(audio,language='en-in')
            print(f"user said:{statement}\n")

        except Exception as e:
            speak("Pardon me, please say that again")
            return "None"
        return statement

print("Loading your AI personal assistant luca")
speak("Loading your AI personal assistant luca")
```

```python
def tellDay():
      
    # This function is for telling the day of the week
    day = datetime.datetime.today().weekday() + 1
      
    Day_dict = {1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday', 7: 'Sunday'}
      
    if day in Day_dict.keys():
        day_of_the_week = Day_dict[day]
        print(day_of_the_week)
        speak("The day is " + day_of_the_week)
```

```python
def tellTime():
      
    # This method will give the time
    time = str(datetime.datetime.now())
    
    # In this the time will be printed along with the date
    
    print(time)
    hour = time[11:13]
    min = time[14:16]
    speak("The time is sir" + hour + "Hours and" + min + "Minutes") 
```
